## Data Collection code

Here we will query similar artists and create a csv

good.csv - contains pairs of artists where both artists are in the artist list

bad.csv  - contains all other pairs

```python
artist_list = full_artist_list

fieldnames = ['main_artist', 'sim_artist', 'score']    
with open('good.csv', 'w') as goodfile:
    gwriter = csv.DictWriter(goodfile, fieldnames=fieldnames)
    gwriter.writeheader()
    with open('bad.csv', 'w') as badfile:
        bwriter = csv.DictWriter(badfile, fieldnames=fieldnames)
        bwriter.writeheader()
        for artist in artist_list:
            modified = artist.replace("-", "%20")
            url = 'http://www.similar-artist.com/similarto/artist/{}/'.format(modified)
            try:
                html_doc = urllib.request.urlopen(url)
                soup = BeautifulSoup(html_doc, 'html.parser')
                next_soup = soup.body.contents[3]
                arr = next_soup.prettify().splitlines()
                for i in range(len(arr)):
                    if "Match Index" in arr[i]:
                        match_str = arr[i].strip().split()[2]
                        sim_artist = arr[i-3].strip().lower().replace(' ', '-')
                        if sim_artist in complete_artist_set:
                            gwriter.writerow({'main_artist' : artist, 'sim_artist' : sim_artist, 'score' : match_str})
                        else:
                            bwriter.writerow({'main_artist' : artist, 'sim_artist' : sim_artist, 'score' : match_str})
                print(artist)
            except urllib.error.URLError as e:
                print(artist, e.reason)
    
# for artist in artist_list:
#     modified = artist.replace("-", "%20")
#     url = 'http://www.similar-artist.com/similarto/artist/{}/'.format(modified)
#     try:
#         html_doc = urllib.request.urlopen(url)
#         soup = BeautifulSoup(html_doc, 'html.parser')21
#         next_soup = soup.body.contents[3]
#         arr = next_soup.prettify().splitlines()
#         for i in range(len(arr)):
#             if "Match Index" in arr[i]:
#                 match_str = arr[i].strip().split()[2]
#                 sim_artist = arr[i-3].strip().lower().replace(' ', '-')
#                 print(artist, match_str, sim_artist)
#                 break
#     except urllib.error.URLError as e:
#         print(e.reason)

```
