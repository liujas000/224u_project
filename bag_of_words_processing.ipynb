```python
from util import load_complete

from collections import defaultdict
import string
import csv
```

```python
'''
Reads in all lyrics to obtain complete set of unique terms and their document frequencies

'''

filename = 'data/lyrics_380k.csv'

unique_terms = set([]) # set of all unique terms in all lyrics
doc_freqs = defaultdict(int) # dictionary of unique terms to number of documents term appears in
num_docs = 0

# Load data
data = load_complete(filename)
```

```python
table = str.maketrans({key: None for key in string.punctuation}) # needed for parsing lyric string to remove punctuation

# For each lyric, add all unique terms to set and update document frequency
for artist in data:
    for lyric in data[artist]['lyrics']:
        
        parsed_lyric = lyric.translate(table) # remove punctuation
        parsed_lyric = ' '.join(parsed_lyric.splitlines()).lower() # remove new lines and make lowercase
        parsed_lyric = parsed_lyric.split() # list of all terms in lyric
        lyric_terms = set(parsed_lyric) # only unique terms in lyric
        lyric_terms = list(lyric_terms)
        
        # Add unique terms from lyrics to overall set and update document frequency
        for term in lyric_terms:
            unique_terms.add(term)
            doc_freqs[term] += 1
        
        num_docs += 1
        
        if num_docs % 10000 == 0:
            print(num_docs)

unique_terms = list(unique_terms)
print('Num Terms:', len(unique_terms))
print('Num Docs:', num_docs)
            
# Divide counts by number of docs to get frequency
for term in unique_terms:
    doc_freqs[term] = float(doc_freqs[term]) / num_docs
```

```python
for term in unique_terms:
    print(term)
```

```python
# Write all terms and their doc frequency to output file
unique_terms.sort() # alphabetize

output = open('terms.tsv', 'w')

for term in unique_terms:
    freq = doc_freqs[term]
#     print(term, freq)
    output.write(term + '\t' + str(freq) + '\n')

output.close()
```
